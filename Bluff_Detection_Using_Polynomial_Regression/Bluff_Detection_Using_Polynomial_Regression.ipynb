# An Machine learning model that can help HR team by detecting the bluff of an candidate regarding the negotiation salary wrt the years of experience.

# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv("Salary_Data.csv")
```

```python
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

# Linear Regression Implementation

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

# Polynomial Linear Regression Implementation

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree = 7)
```

```python
X_poly = pr.fit_transform(X)
print(X_poly)
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(X_poly, y)
```

# Visualization

## Visualizing the results of Linear Regression

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lr.predict(X), color = "blue")
plt.title("Bluff Detection (Linear Regression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Visualizing the results of Ploynomial Linear Regression

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lr2.predict(pr.fit_transform(X)), color = "blue")
plt.title("Bluff Detection (Polynomial Linear Regression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

# Prediction

## Prediction with Linear Regression Model

```python
lr_pred = lr.predict([[7.75]])
print(lr_pred)
```

## Prediction with Polynomial Linear Regression Model

```python
pr_pred = lr2.predict(pr.fit_transform([[7.75]]))
print(pr_pred)
```

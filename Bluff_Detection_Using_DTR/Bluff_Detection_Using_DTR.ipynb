# An Machine learning model that can help HR team by detecting the bluff of an candidate regarding the negotiation salary wrt the years of experience.

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Salary_Data.csv")
```

```python
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Decision Tree Regression Implementation

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state = 0)
```

```python
dtr.fit(X, y)
```

## Decision Tree Regression Visualization

```python
plt.scatter(X, y, color = "red")
plt.plot(X, dtr.predict(X), color = "blue")
plt.title("Bluff Detection (Decision Tree Regression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Decision Tree Regression Visualization with Improved Resolution

```python
X_grid = np.arange(min(X), max(X), 0.1) # 0.1 : Resolution
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
plt.scatter(X, y, color = "red")
plt.plot(X_grid, dtr.predict(X_grid), color = "blue")
plt.title("Bluff Detection (Decision Tree Regression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Prediction with Decision Tree Regression

```python
pred = dtr.predict(np.reshape(7.5, (1, 1)))
print(pred)
```

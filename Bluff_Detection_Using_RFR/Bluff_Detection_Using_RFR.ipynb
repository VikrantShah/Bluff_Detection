# An Machine learning model that can help HR team by detecting the bluff of an candidate regarding the negotiation salary wrt the years of experience.

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Salary_Data.csv")
```

```python
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Random Forest Regression Implementation

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators = 15, random_state = 0)
```

```python
rfr.fit(X, y)
```

## Random Forest Regression Visualization

```python
plt.scatter(X, y, color = "red")
plt.plot(X, rfr.predict(X), color = "blue")
plt.title("Bluff Detection (Random Forest Regressor)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Random Forest Regression Visualization with Imporoved Resolution

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
plt.scatter(X, y, color = "red")
plt.plot(X_grid, rfr.predict(X_grid), color = "blue")
plt.title("Bluff Detection (Random Forest Regression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Prediction with Random Forest Regression

```python
pred = rfr.predict(np.reshape(7.5, (1, 1)))
print(pred)
```
